## EX2

```python
x_train = np.linspace(0,1,10)
y_train = [np.sin(2 * np.pi * i) + np.random.normal(scale= 0.25) for i in x_train]

x_test = np.linspace(0, 1, 100)
y_test = [np.sin(2 * np.pi * i) for i in x_test]

fig = plt.figure(figsize = (12,6))
plt.scatter(x_train, y_train)
plt.plot(x_test, y_test, c="g")
plt.show()
```

```python
model_0 = Linear_Regression.LinearRegression()
x_poly = Polynomial_Feature.Polynomial_Feature(0)
x_poly_train = x_poly.fit_transform(x_train)
x_poly_test = x_poly.fit_transform(x_test)
model_0.fit(x_poly_train, y_train)
```

```python
fig = plt.figure(figsize = (12,6))
plt.scatter(x_train, y_train)
plt.plot(x_test, y_test, c="g")
y_pred = model_0.predict(x_poly_test)
plt.plot(x_test , y_pred, c= "r")
```

#### With M = 0: model can be visualized as a line parallel with Ox axes. Just a few amount of points this line can fit with the test set. This model might consider as underfitting

```python
model_1 = Linear_Regression.LinearRegression()
x_poly = Polynomial_Feature.Polynomial_Feature(1)
x_poly_train = x_poly.fit_transform(x_train)
x_poly_test = x_poly.fit_transform(x_test)
model_1.fit(x_poly_train, y_train)
```

```python
fig = plt.figure(figsize = (12,6))
plt.scatter(x_train, y_train)
plt.plot(x_test, y_test, c="g")
y_pred = model_1.predict(x_poly_test)
plt.plot(x_test , y_pred, c= "r")
```

#### With M = 1: model can be visualized as a down slope line with . Just a few amount of points this line can fit with the test set but this model still better than with M = 0. This model might consider as underfitting
