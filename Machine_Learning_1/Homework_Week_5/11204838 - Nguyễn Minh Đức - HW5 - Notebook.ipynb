```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
import Logistic_Regression
```

```python
df = pd.read_csv('dataset.csv')
X = df.iloc[:, :-1]
y = df.iloc[:,-1]

clf = Logistic_Regression.LogisticRegression()
clf.fit(X, y)
model_theta = clf.get_theta()
model_theta
```

```python
X_approved = df[df["Cho vay"] == 1]
X_denied = df[df["Cho vay"] == 0]


fig , ax = plt.subplots(1, 1)
ax.scatter(X_approved.iloc[:, 0], X_approved.iloc[:, 1], c= "red", label = "Approved")
ax.scatter(X_denied.iloc[:, 0], X_denied.iloc[:, 1], c= "blue", label = "Denied")
ax.legend(loc= 1)
ax.set_xlabel('Salary (million)')
ax.set_ylabel('Experience (years)')

t = 0.5
ax.plot((4,10), (-(model_theta[0] + model_theta[1] * 4 + np.log(1/t-1)) / model_theta[2] , -(model_theta[0] + model_theta[1] * 10 + np.log(1/t-1) )/ model_theta[2]))

plt.show()
```
